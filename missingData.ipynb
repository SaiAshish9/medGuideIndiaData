```python
arr=[]
```

```python
def convert(x,a):
        keys=x[:][:1].values.tolist()[0]
        for i in range(1,a-1):
                c=x[:][i:].values.tolist()[0]
                obj={}
                for i in range(0,len(keys)):
                        obj[keys[i]]=c[i]  
                arr.append(obj)
        print('{} out of 1327'.format(len(arr)),'{} rows extracted'.format(len(arr)))
```

```python
for y in urls:
    z=np.array(pd.read_html(y))[12][:][:51]
    a=len(pd.DataFrame(z).count(axis=1).tolist())+1
    z=np.array(pd.read_html(y))[12][:][:a]
    z=z[z.columns[0:10]]
    convert(z,a)
```
