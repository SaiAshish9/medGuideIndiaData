```python
def convert(x,y):
        keys=x[:][:1].values.tolist()[0]
        for i in range(1,51):
                c=x[:][i:].values.tolist()[0]
                obj={}
                for i in range(0,len(keys)):
                        if i == len(keys)-1:
                                arr=[]
                                html_page = urlopen(y)
                                soup = BeautifulSoup(html_page)
                                for link in soup.findAll('a'):
                                        if link.get('onclick'):
                                                arr.append('http://www.medguideindia.com/'+link.get('onclick').split("'")[1].split("'")[0])
                                a=np.array(pd.read_html(arr[0]))[1][:2]
                                ingredients={}
                                ingredients[a[0][0]]=a[0][1]
                                ingredients[a[1][0]]=a[1][1]
                                obj[keys[i]]=ingredients
                        else:
                            obj[keys[i]]=c[i]
                array.append(obj)
        print('{} out of 2187'.format(len(arr)/50),'{} rows extracted'.format(len(arr)))
```
